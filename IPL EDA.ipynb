### Comparision of Two Team

```python
def compare_two_team(team1,team2):
    compare_df=df[((df["Team1"]==team1) & (df["Team2"]==team2)) | 
              ((df["Team1"]==team2) & (df["Team2"]==team1))][['Team1','Team2','WinningTeam','WonBy','Player_of_Match']]
    team1_name=compare_df["WinningTeam"].value_counts().index[0]
    team2_name=compare_df["WinningTeam"].value_counts().index[1]
    team1_win=compare_df["WinningTeam"].value_counts().values[0]
    team2_win=compare_df["WinningTeam"].value_counts().values[1]
    
    team1_bowling_first=len(compare_df[(compare_df['WinningTeam'] == team1) & (compare_df['WonBy']=='Wickets')])
    team1_batting_first=len(compare_df[(compare_df['WinningTeam'] == team1) & (compare_df['WonBy']=='Runs')])
    team2_bowling_first=len(compare_df[(compare_df['WinningTeam'] == team2) & (compare_df['WonBy']=='Wickets')])
    team2_batting_first=len(compare_df[(compare_df['WinningTeam'] == team2) & (compare_df['WonBy']=='Runs')])
    
    Player_of_the_match=compare_df['Player_of_Match'].value_counts().index[0]
    no_of_time_potm=compare_df['Player_of_Match'].value_counts().values[0]
  
    print(f'\t\tTotal Match Played : {len(compare_df)}')
                                                             
    print(f'Highest Player of the Match : {Player_of_the_match}({no_of_time_potm})')
                                                             
    data={
        "Total Win" : [team1_batting_first+team1_bowling_first,team2_batting_first+team2_bowling_first],
        'Batting First' : [team1_batting_first,team2_batting_first],
        'Bowling First' : [team1_bowling_first,team2_bowling_first]
    }
    return pd.DataFrame(data,index=(team1,team2))
```

### Q)Most successful batting pair in IPL

```python
def func(x):
    return '-'.join(list(np.sort(x.values)))
balls["batter-pair"] = balls[["batter", "non-striker"]].apply(func, axis = 1) 
temp_df = balls.groupby("batter-pair").agg(
{
    "total_run": "sum",
    "ballnumber": "count",
    "isWicketDelivery": "sum"
}).reset_index()
temp_df["Batsman 1"] = temp_df["batter-pair"].apply(lambda x: x.split("-")[0])
temp_df["Batsman 2"] = temp_df["batter-pair"].apply(lambda x: x.split("-")[1])
temp_df.rename(columns = {"total_run": "Runs"}, inplace=True)
temp_df["StrikeRate"] = round(temp_df["Runs"] / temp_df["ballnumber"] * 100,2)
temp_df["Avg"] = round(temp_df["Runs"] / temp_df["isWicketDelivery"],2)
temp_df.sort_values("Runs", ascending = False, inplace=  True)
temp_df[["Batsman 1", "Batsman 2", "Runs", "Avg", "StrikeRate"]].head(10)
```

### Q)Batsman record season wise.

```python
batterdf = seasondf.copy()
batterdf["IsBatsmanBall"] = batterdf["extra_type"].apply(lambda x: 1 if x != "wides" else 0)

def bat_record_season(batsman):
    bdf = batterdf[batterdf.batter == batsman].copy()
    bdf["IsBatsmanOut"] = bdf.batter == bdf.player_out
    df = bdf.groupby(["Season", "ID"], as_index = False)[["batsman_run", "IsBatsmanBall", "IsBatsmanOut"]].sum()
    innings = df.groupby("Season").ID.count()
    df = df.groupby("Season").agg(
    {
        "batsman_run":["sum", "max"],
        "IsBatsmanBall": "sum",
        "IsBatsmanOut": "sum"
    })
    df["Innings"]= innings
    df["TotalRuns"] = df[("batsman_run", "sum")]
    df["Avg"] = round(df["TotalRuns"] / df[("IsBatsmanOut", "sum")],2)
    df["HighestScore"] = df[("batsman_run", "max")]
    df["StrikeRate"] = round(df["TotalRuns"] / df[("IsBatsmanBall", "sum")] * 100,2)
    print(f'\t\t\t{batsman}')
    return df.drop(columns = ["batsman_run", "IsBatsmanBall", "IsBatsmanOut"])
    
bat_record_season("V Kohli")
```

### Q)Most Economical Bowler (bowl at least 600 balls)

```python
bowler_seasondf=seasondf.copy()
bowler_seasondf["BowlerRun"] = bowler_seasondf.extra_type.apply(lambda x: 0 if x in ["legbyes", "byes"] else 1) * bowler_seasondf["total_run"]
bowler_seasondf["IsLegalBall"] = bowler_seasondf.extra_type.apply(lambda x: 0 if x in ["wides", "noballs"] else 1)

eco=bowler_seasondf.groupby('bowler').agg({'BowlerRun':'sum','IsLegalBall':'count'}).sort_values(ascending=False,by='BowlerRun')
eco['Economy']=eco["BowlerRun"] /eco["IsLegalBall"] * 6
eco[eco['IsLegalBall'] >=600].sort_values(by='Economy')[['Economy']].head(10)
```

```python
bowler_seasondf=seasondf.copy()
bowler_seasondf["BowlerRun"] = bowler_seasondf.extra_type.apply(lambda x: 0 if x in ["legbyes", "byes"] else 1) * bowler_seasondf["total_run"]
bowler_seasondf["IsLegalBall"] = bowler_seasondf.extra_type.apply(lambda x: 0 if x in ["wides", "noballs"] else 1)
bowler_seasondf['IsBowlerWicket'] = bowler_seasondf.kind.apply(lambda x : 1 if x in ["caught", 'caught and bowled', 'bowled', 'stumped','lbw', 'hit wicket'] else 0)
pcapdf = bowler_seasondf.groupby(["Season", "bowler"], as_index = False)[["IsBowlerWicket", "BowlerRun", "IsLegalBall"]].sum()
pcapdf["Economy"] = round(pcapdf["BowlerRun"] / pcapdf["IsLegalBall"] * 6,2)
pcapdf_=pcapdf.sort_values(["IsBowlerWicket", "Economy"], ascending = [False, True]).drop_duplicates("Season", keep = "first").sort_values("Season")
pcapdf_=pcapdf_[['Season','bowler','IsBowlerWicket','Economy']]
pcapdf_.columns=['Season','Bowler','Wicket','Economy']
pcapdf_
```
